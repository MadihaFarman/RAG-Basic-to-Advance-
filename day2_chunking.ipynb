## Fixed Sized Chunking

```python
def fixed_sized_chunks(text, chunk_size=500, overlap=50):
    chunks = []
    start = 0
    while (start<len(text)):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap
    return chunks
```

## Semantic Embeddings 

```python
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize
import os
from huggingface_hub import InferenceClient
from dotenv import load_dotenv

load_dotenv()

# Initialize the client with the inference provider
client = InferenceClient(
    provider="hf-inference",
    api_key=os.environ["HF_TOKEN"],
)

model_id = "sentence-transformers/all-MiniLM-L6-v2"
document = (
"The university offers a BS in Computer Science. The program runs "
"for four years across eight semesters. Admission requires a CNIC "
"and academic transcripts. "
"The cafeteria serves lunch from 12pm to 3pm daily. Vegetarian "
"options are available every day. "
"The library is open until 10pm on weekdays."
)
sentences = sent_tokenize(document)
embeddings = client.feature_extraction(text=document,model=model_id,normalize=True)
# Compute similarity between EVERY pair of consecutive sentences
similarities = []
for i in range(len(embeddings) - 1):
 sim = np.dot(embeddings[i], embeddings[i + 1])
 similarities.append(sim)
 print(f"Sentence {i} <-> {i+1}: similarity = {sim:.3f}")
# Find where similarity drops below a threshold -- that's a topic boundary
threshold = 0.5
boundaries = [i for i, sim in enumerate(similarities) if sim < threshold]
print("\nTopic boundaries detected after sentence indices:", boundaries)
# Build chunks by grouping sentences BETWEEN detected boundaries
def build_chunks(sentences, boundaries):
 chunks = []
 start = 0
 for b in boundaries:
  chunks.append(" ".join(sentences[start:b+1]))
  start = b + 1
 chunks.append(" ".join(sentences[start:]))
 return chunks
semantic_chunks = build_chunks(sentences, boundaries)
for i, c in enumerate(semantic_chunks):
 print(f"\n--- Semantic chunk {i} ---\n{c}")

# Expected result: one chunk about the CS program/admission, a separate
# chunk about the cafeteria, and a separate chunk about the library --
# because the embedding similarity drops noticeably between these topics
# even though there was no explicit paragraph break in the raw text 
```

## Parent-child chunking

```python
# Conceptual implementation of parent-child chunking
from langchain_text_splitters import RecursiveCharacterTextSplitter
document = """
Section 3: Refund Policy
Students who withdraw within the first two weeks of the semester
are eligible for a full refund of tuition fees. Withdrawals between
weeks three and six receive a 50 percent refund. No refunds are
issued after week six under any circumstances, except in cases of
documented medical emergency, which are reviewed on a case-by-case
basis by the Dean's office.
"""
# Step 1: split into large PARENT chunks (full sections, e.g. 800 chars)
parent_splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=0)
parent_chunks = parent_splitter.split_text(document)
# Step 2: for each parent, split into small CHILD chunks (e.g. 150 chars)
child_splitter = RecursiveCharacterTextSplitter(chunk_size=150, chunk_overlap=20)
parent_store = {}# parent_id -> full parent text
child_to_parent = {}# child text -> parent_id (in production, use a real DB)
for p_idx, parent_text in enumerate(parent_chunks):
 parent_id = f"parent_{p_idx}"
 parent_store[parent_id] = parent_text
 children = child_splitter.split_text(parent_text)
 for child_text in children:
    child_to_parent[child_text] = parent_id
# At INDEXING time: embed and store only the CHILD chunks in the vector DB
print("Children to embed and index:")
for child in child_to_parent:
 print(" -", child[:60], "...")
# At RETRIEVAL time: vector search finds a relevant CHILD chunk,
# e.g. "Withdrawals between weeks three and six receive a 50 percent refund."
matched_child = list(child_to_parent.keys())[1]
matched_parent_id = child_to_parent[matched_child]
full_context = parent_store[matched_parent_id]
print("\nMatched (small, precise) child chunk:")
print(matched_child)
print("\nRetrieved (large, complete) parent context sent to the LLM:")
print(full_context)
# The LLM receives the FULL refund policy section -- including the full
# refund and medical emergency exception clauses -- not just the isolated
# sentence about the 50% refund, which alone could mislead a student into
# missing the medical emergency exception. 
```
