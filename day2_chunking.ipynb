## Fixed Sized Chunking

```python
def fixed_sized_chunks(text, chunk_size=500, overlap=50):
    chunks = []
    start = 0
    while (start<len(text)):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap
    return chunks
```

## Recursive Character Splitting

fixes the "cuts mid-sentence" problem of fixed-size chunking while staying simple and fast.
The core idea: try to split on the most meaningful separator first, and only fall back to a less meaningful
separator if a chunk is still too large.
