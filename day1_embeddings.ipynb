## What a single embedding vector looks like

```python
import os
from huggingface_hub import InferenceClient
from dotenv import load_dotenv

load_dotenv()

# Initialize the client with the inference provider
client = InferenceClient(
    provider="hf-inference",
    api_key=os.environ["HF_TOKEN"],
)

# Define the text and the target model ID
model_id = "sentence-transformers/all-MiniLM-L6-v2"
text = "RAG systems retrieve relevant documents before generating an answer"

# Call feature_extraction to get the remote embedding
embedding = client.feature_extraction(text, model=model_id)

print(type(embedding))
# Output: <class 'numpy.ndarray'>

print(embedding.shape)
# Output: (384,) -- 384 floating point numbers

print(embedding[:8])
# Output: array([-0.0234..., ...], dtype=float32)
```

## Similarity Metrics

```python
import numpy as np 
sentences = [
"A doctor examined the patient",
"A physician checked on the patient",
"The weather is sunny today",
]

# without noralization
embedding_1 = client.feature_extraction(sentences,model=model_id)
print("Magnitude of each Vector before normalization:",np.linalg.norm(embedding_1,axis=1))

```

```python
# with noralization
embedding_2 = client.feature_extraction(sentences,model=model_id,normalize=True)
print("Magnitude of each Vector after normalization:",np.linalg.norm(embedding_1, axis=1))
```

```python
# cosine similarity via dot product (bcz magnitude is normalized therefore cosine sim = dot product)

sim_dr_physician = np.dot(embedding_2[0],embedding_2[1])
sim_dr_weather = np.dot(embedding_2[0],embedding_2[2])
print("dot/cosine similarity b/w dr and phsyician : ",sim_dr_physician)
print("dot/cosine similarity b/w dr and weather : ",sim_dr_weather)
```
